# Handling missing values

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
si.fit(X_train[:,:])
X_train[:,:]=si.transform(X_train[:,:])
si.fit(X_test[:,:])
X_test[:,:]=si.transform(X_test[:,:])
```
